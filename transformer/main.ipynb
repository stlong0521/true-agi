# How to run this notebook on Google Colab from local IDE
Prerequisites
* Install Google Colab extension in IDE (VS Code / Antigravity)
* Authorize Google Colab to access Github repos in Google Colab browser UI (settings)

Steps to run
1. Start this notebook in local IDE, select/connect to the right kernel (CPU/GPU/etc.); at first time, need to sign in with Google account
2. Run the notebook cell by cell in local IDE
3. With any local code changes (outside this notebook), always push to Github and then pull from Github in this notebook to sync code changes

What to do after
* Explicitly save this file (Command + S) before adding changes of this file to source control (git)

## Clone repo, set up environment and dependencies (via uv)

```python
%%bash
# Check if the repo dir already exists
DIR="/content/true-agi"
if [ -d "$DIR" ]; then
    cd "$DIR"
    git pull
else
    git clone https://github.com/stlong0521/true-agi.git
    cd "$DIR"
fi

# Confirm the current working dir
pwd

# Check if uv is installed or not
if command -v uv &> /dev/null; then
    uv sync
else
    curl -LsSf https://astral.sh/uv/install.sh | sh
    uv sync
fi
```
